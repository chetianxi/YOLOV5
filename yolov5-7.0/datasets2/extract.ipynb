```python
video = cv2.VideoCapture("./zhilang.mp4")
num = 0
save_step = 30
while True:
    ret, frame = video.read()
    if not ret:
        break
    num += 1
    if num % save_step == 0:
        cv2.imwrite("./images/" + str(num) + ".jpg", frame)
```
